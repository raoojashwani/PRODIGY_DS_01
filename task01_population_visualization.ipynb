```python
from google.colab import files
uploaded = files.upload()

# TASK 01 – Distribution Visualization

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_6298016.csv", skiprows=4)

# Select recent year and top 10 countries by population
df_latest = df[['Country Name', '2022']]
df_latest = df_latest.dropna()
df_latest = df_latest.sort_values('2022', ascending=False).head(10)

# Plot Bar Chart
plt.figure(figsize=(12, 6))
plt.bar(df_latest['Country Name'], df_latest['2022'] / 1e6, color='teal')
plt.title("Top 10 Most Populated Countries in 2022")
plt.ylabel("Population (in Millions)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
